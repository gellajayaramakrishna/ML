```python
#3
import numpy as np
import time

def pass_median(A):
    A.sort()
    n = len(A)
    if n % 2 == 1:
        return A[n // 2]
    else:
        return (A[n // 2 - 1] + A[n // 2]) / 2

def quickselect_median(A, k):
    if len(A) == 1:
        return A[0]
    pivot = A[len(A)//2]
    left = [x for x in A if x < pivot]
    middle = [x for x in A if x == pivot]
    right = [x for x in A if x > pivot]
    if k < len(left):
        return quickselect_median(left, k)
    elif k < len(left) + len(middle):
        return middle[0]
    else:
        return quickselect_median(right, k - len(left) - len(middle))

def numpy_median(A):
    return np.median(A)

A = [12, 11, 9, 34, 0, 96]

start_time = time.time()
print("Pass Median:", pass_median(A.copy()))
print("Time taken:", time.time() - start_time)

start_time = time.time()
print("Quick Select Median:", quickselect_median(A.copy(), len(A) // 2))
print("Time taken:", time.time() - start_time)

start_time = time.time()
print("NumPy Median:", numpy_median(A))
print("Time taken:", time.time() - start_time)
```
